# 1D Tensor

```python
import torch

# create a 1D tensor
tensor_data = [1, 2, 5, 7, 9]
tensor = torch.tensor(tensor_data, dtype = torch.float32)

# Display the tensor
print("Original Tensor: ", tensor)
```

```python
# Now manipulate the tensor
# Example: Power each element

tensor = tensor ** 3

# Display the tensor
print("Manipulated Tensor: ", tensor)
```

# 2D Tensor

```python
# create a tensor
tensor_data = [[1,3,5,7], [2,4,6,8], [1,2,3,4]]
tensor = torch.tensor(tensor_data, dtype = torch.float32)

# Display the created tensor
print("Original Tensor: ")
print(tensor)
```

```python
# Manipulate the tensor
# Example power of each element
# tensor = tensor ** 2
tensor *= 2

# Display the manipulated Tensor
# print("Manipulated Tensor: ", tensor)
print("Manipulated tensor: ")
print(tensor)
```

<!-- break -->

# visualization

```python
import matplotlib.pyplot as plt
plt.imshow(tensor, cmap = 'plasma', interpolation ='nearest')
plt.title("Visualization of Manipulated Tensor")
plt.colorbar()
plt.show()
```

# Now work with image

```python
import torch
import matplotlib.pyplot as plt
from torchvision import transforms
from PIL import Image

# Load an image using PIL (PIL == PYTHON IMAGING LIBRARY)
image_path = '/content/drive/MyDrive/forest.jpg'
image = Image.open(image_path)

# Display the original image
plt.imshow(image)
plt.title("Original Image")
plt.show()

# Convert the PIL image to a PyTorch tensor
image_transform = transforms.ToTensor()
tensor_image = image_transform(image)

# Display the PyTorch tensor version of the image
# In PIL the default order is (H, W, C) and in PyTorch the order is (C, H, W)
plt.imshow(tensor_image.permute(1, 2, 0))   # (C, H, W)
plt.title("Tensor Version of the Image")
plt.show()
```

# 2D Image

```python
import torch
import matplotlib.pyplot as plt
from torchvision import transforms
from PIL import Image

# Load a grayscale image using PIL (Python Imaging Library)
image_path = '/content/drive/MyDrive/Picture/gray_baby.jpg'
# image_path = '/content/drive/MyDrive/Picture/gray_Mona_Lisa.jpg'

image = Image.open(image_path)

# Display the original image
plt.imshow(image, cmap = 'gray')
plt.title("Original Image")
plt.show()

# Convert the PIL image to a PyTorch tensor
image_transform = transforms.ToTensor()
tensor_image = image_transform(image)

# Display the PyTorch version of the images
plt.imshow(tensor_image[0], cmap = 'gray')
plt.title("Tensor version of the images")
plt.show()

# print the PyTorch tensor
print("PyTorch tensor: ")
print(tensor_image)   # Display (channel, Height, Width)
```

**By default, every images consider as a RGB images in PyTorch. Here is our picture is Grayscale images but the PyTorch consider it into a 3D/RGB images. So, we got here the channel is 3.... **
